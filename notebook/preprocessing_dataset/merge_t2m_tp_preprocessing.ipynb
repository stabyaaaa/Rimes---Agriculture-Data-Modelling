```python
import pandas as pd
import numpy as np

# --- 0ï¸âƒ£ Ensure your dataframe `df` is loaded ---
# df = pd.read_csv('your_data.csv')  # Example

# 1ï¸âƒ£ Ensure datetime column
df['date'] = pd.to_datetime(df['date'])

# 2ï¸âƒ£ Define full daily range for the study period
full_dates = pd.date_range(start='1940-01-01', end='2025-10-31', freq='D')

# 3ï¸âƒ£ Function to fill missing calendar days per (lat, lon)
def fill_missing_dates(group):
    group = group.set_index('date').reindex(full_dates)
    group.index.name = 'date'
    
    # Restore fixed identifiers
    group['latitude'] = group['latitude'].ffill().bfill().iloc[0]
    group['longitude'] = group['longitude'].ffill().bfill().iloc[0]
    
    # Fill temporal parts
    group['year'] = group.index.year
    group['month'] = group.index.month
    group['day'] = group.index.day

    # --- Variable-specific filling ---
    # Temperature: interpolate
    if 't2m_C' in group.columns:
        group['t2m_C'] = group['t2m_C'].interpolate(limit_direction='both')
    
    # Precipitation: fill missing with 0
    if 'tp_mm' in group.columns:
        group['tp_mm'] = group['tp_mm'].fillna(0)
    
    # Other numeric columns: interpolate
    numeric_cols = group.select_dtypes(include=[np.number]).columns.difference(['t2m_C','tp_mm'])
    group[numeric_cols] = group[numeric_cols].interpolate(limit_direction='both')

    # Flags / categorical columns
    for col in group.columns:
        if group[col].dtype == 'object':
            group[col] = group[col].fillna(method='ffill').fillna(method='bfill')
        if group[col].dtype == 'bool' or 'flag' in col.lower():
            group[col] = group[col].fillna(0).astype(int)

    return group.reset_index()

# 4ï¸âƒ£ Apply filling per location
df_filled = df.groupby(['latitude', 'longitude'], group_keys=False).apply(fill_missing_dates)
df_filled = df_filled.sort_values(['latitude','longitude','date']).reset_index(drop=True)

print("âœ… Missing dates filled successfully!")

# 5ï¸âƒ£ Compute heat_proxy and future heat flags per location
def compute_heat_flags(group):
    # 90th percentile threshold
    threshold = group['t2m_C'].quantile(0.90)
    group['heat_proxy'] = (group['t2m_C'] >= threshold).astype(int)
    
    # Shift to get future heat days
    group = group.sort_values('date')
    group['heat_next_day']   = group['heat_proxy'].shift(-1, fill_value=0)
    group['heat_next_2days'] = group['heat_proxy'].shift(-2, fill_value=0)
    group['heat_next_3days'] = group['heat_proxy'].shift(-3, fill_value=0)
    
    # Ensure 0/1
    for col in ['heat_next_day','heat_next_2days','heat_next_3days']:
        group[col] = group[col].clip(0,1).astype(int)
    
    return group

# 6ï¸âƒ£ Apply heat computation
df_final = df_filled.groupby(['latitude', 'longitude'], group_keys=False).apply(compute_heat_flags)
df_final = df_final.sort_values(['latitude','longitude','date']).reset_index(drop=True)

# 7ï¸âƒ£ Optional: check last 10 rows
print(df_final[['date','latitude','longitude','t2m_C','tp_mm','heat_proxy','heat_next_day','heat_next_2days','heat_next_3days']].tail(10))
```

Cumulative precipitation per location

Key points:

groupby(['latitude','longitude']) ensures you only sum precipitation within the same area.

rolling(7) gives a moving 7-day sum, not total across the dataset.

reset_index(level=[0,1], drop=True) avoids the MultiIndex error when assigning back to df

```python
# 7-day cumulative precipitation per location
df['tp_7d_cum'] = (
    df.groupby(['latitude','longitude'])['tp_mm']
      .rolling(7, min_periods=1)  # 7-day window
      .sum()
      .reset_index(level=[0,1], drop=True)  # match original df index
)

# 14-day cumulative
df['tp_14d_cum'] = (
    df.groupby(['latitude','longitude'])['tp_mm']
      .rolling(14, min_periods=1)
      .sum()
      .reset_index(level=[0,1], drop=True)
)
```

Lag features are previous values of a variable added as new columns, giving your model temporal context. For precipitation (tp_mm), this helps the model â€œseeâ€ what happened in previous days.

Why lag features are important

Rainfall is temporally correlated; past precipitation often influences current crop stress or soil moisture.

Helps the model detect:

Drying trends â†’ drought risk

Successive heavy rains â†’ waterlogging risk

Patterns for heat/drought interaction

```python
import pandas as pd

# Assume df has columns: ['latitude','longitude','date','tp_mm']
df = df.sort_values(['latitude','longitude','date'])

# Lag 1-day, 3-day, 7-day
df['tp_lag1'] = df.groupby(['latitude','longitude'])['tp_mm'].shift(1)
df['tp_lag3'] = df.groupby(['latitude','longitude'])['tp_mm'].shift(3)
df['tp_lag7'] = df.groupby(['latitude','longitude'])['tp_mm'].shift(7)
```

This will tell the fraction of rows where the forward-looking heat values match your heat_proxy shifts.

If any of the fractions are less than 1.0, it means some values may be off â€” probably due to thresholding or grouping.

```python
import pandas as pd

# Assuming your DataFrame is called df
# Sort by location and date just to be safe
df = df.sort_values(['latitude', 'longitude', 'date'])

# # Shift heat_proxy to get next days
# df['check_next_day'] = df.groupby(['latitude', 'longitude'])['heat_proxy'].shift(-1)
# df['check_next_2days'] = df.groupby(['latitude', 'longitude'])['heat_proxy'].shift(-2)
# df['check_next_3days'] = df.groupby(['latitude', 'longitude'])['heat_proxy'].shift(-3)

# Compare with your existing columns
df['next_day_match'] = df['heat_next_day'] == df['check_next_day']
df['next_2days_match'] = df['heat_next_2days'] == df['check_next_2days']
df['next_3days_match'] = df['heat_next_3days'] == df['check_next_3days']

# Summary
summary = df[['next_day_match', 'next_2days_match', 'next_3days_match']].mean()
print("Fraction of rows matching:")
print(summary)
```
