GPT : The reason we calculate anomalies is:

Removes the long-term mean → focuses on deviations rather than absolute temperature.

Makes comparisons easier across time or locations → you can see which years or regions are warmer or colder relative to their typical climate.

Highlights extremes and trends → a +3°C anomaly is meaningful even if one location is naturally colder than another.

So while the shape of the time series looks similar, anomalies are relative to the baseline, which is important for climate studies, crop stress analysis, or detecting warming/cooling trends.
