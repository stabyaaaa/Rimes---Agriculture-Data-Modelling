Yearly Max, Min, and Mean Temperature Location

```python
import pandas as pd
import matplotlib.pyplot as plt

df_loc = df[(df["latitude"] == lat_target) & (df["longitude"] == lon_target)].copy()

#yearly min max mean
yearly_temp = df_loc.groupby("year")["t2m_C"].agg(["max", "min", "mean"]).reset_index()

#highest and lowest temperatures
max_temp = yearly_temp["max"].max()
max_year = yearly_temp.loc[yearly_temp["max"] == max_temp, "year"].values[0]

min_temp = yearly_temp["min"].min()
min_year = yearly_temp.loc[yearly_temp["min"] == min_temp, "year"].values[0]


plt.figure(figsize=(14,6))
plt.plot(yearly_temp["year"], yearly_temp["max"], color="red", label="Max Temp")
plt.plot(yearly_temp["year"], yearly_temp["min"], color="blue", label="Min Temp")
plt.plot(yearly_temp["year"], yearly_temp["mean"], color="green", label="Avg Temp")

#marking highest and lowest temperatures
plt.scatter(max_year, max_temp, color="darkred", s=100, label=f"Highest: {max_temp:.1f}°C ({max_year})")
plt.scatter(min_year, min_temp, color="darkblue", s=100, label=f"Lowest: {min_temp:.1f}°C ({min_year})")

plt.title(f"Yearly Max, Min, and Mean Temperature at 9 am at Location: {lat_target}N, {lon_target}E")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Define rice growth stages
def rice_stage(month):
    if month == 6:
        return "Germination"
    elif month == 6 or month == 7:
        return "Seedling"
    elif month == 7 or month == 8:
        return "Vegetative"
    elif month == 8 or month == 9:
        return "Reproductive"
    elif month == 9 or month == 10:
        return "Grain Filling"
    else:
        return "Off-season"

df["growth_stage"] = df["month"].apply(rice_stage)

heat_stage_year = df[df["growth_stage"]!="Off-season"].groupby(["year","growth_stage"])["heat_stress"].sum().unstack().fillna(0)

ax = heat_stage_year.plot(kind="bar", stacked=True, colormap="Reds_r", figsize=(14,6), width=0.8)
ax.set_title("Heat Stress Days per Rice Growth Stage (Yearly)")
ax.set_xlabel("Year")
ax.set_ylabel("Number of Heat Stress Days (>35°C)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Heat Stress Days per Rice Growth Stage (Yearly)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter for Kathmandu approx location
lat_target = 27.75
lon_target = 85.25
df_loc = df[(df["latitude"] == lat_target) & (df["longitude"] == lon_target)].copy()

# Define threshold
heat_threshold = 25
df_loc["heat_stress"] = df_loc["t2m_C"] > heat_threshold

# Define rice growth stages
def rice_stage(month):
    if month == 6:
        return "Germination"
    elif month in [6, 7]:
        return "Seedling"
    elif month in [7, 8]:
        return "Vegetative"
    elif month in [8, 9]:
        return "Reproductive"
    elif month in [9, 10]:
        return "Grain Filling"
    else:
        return "Off-season"

df_loc["growth_stage"] = df_loc["month"].apply(rice_stage)

heat_stage_year = (
    df_loc[df_loc["growth_stage"] != "Off-season"]
    .groupby(["year", "growth_stage"])["heat_stress"]
    .sum()
    .unstack()
    .fillna(0)
)

# Plot
ax = heat_stage_year.plot(
    kind="bar", stacked=True,colormap="Reds_r", figsize=(14, 6), width=0.8
)
ax.set_title("Heat Stress Days per Rice Growth Stage (Yearly) - Kathmandu (27.75, 85.25)")
ax.set_xlabel("Year")
ax.set_ylabel("Number of Heat Stress Days (>35°C)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

GPT : The reason we calculate anomalies is:

Removes the long-term mean → focuses on deviations rather than absolute temperature.

Makes comparisons easier across time or locations → you can see which years or regions are warmer or colder relative to their typical climate.

Highlights extremes and trends → a +3°C anomaly is meaningful even if one location is naturally colder than another.

So while the shape of the time series looks similar, anomalies are relative to the baseline, which is important for climate studies, crop stress analysis, or detecting warming/cooling trends.

```python
long_term_mean = df['t2m_C'].mean()
df['anomaly'] = df['t2m_C'] - long_term_mean
```

```python
import pandas as pd



# Sort by date
df = df.sort_values('date')

# Function to calculate consecutive days
def consecutive_days(stress_series):
    count = 0
    consec = []
    for val in stress_series:
        if val:
            count += 1
        else:
            count = 0
        consec.append(count)
    return consec


df['heat_consec_days'] = consecutive_days(df['heat_stress'])
df['cold_consec_days'] = consecutive_days(df['cold_stress'])

df['heatwave_flag'] = df['heat_consec_days'] >= 3
df['coldsnap_flag'] = df['cold_consec_days'] >= 3

df[['date','t2m_C','heat_stress','heat_consec_days','heatwave_flag',
    'cold_stress','cold_consec_days','coldsnap_flag']].head(20)
```
